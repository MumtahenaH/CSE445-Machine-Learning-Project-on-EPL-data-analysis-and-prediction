```python
# Calculate games played for each player
games_played_df = all_premier_players_apps.groupby('player_id')['appearance_id'].count().reset_index()
games_played_df.rename(columns={'appearance_id': 'games_played'}, inplace=True)

# Merge with premier_clubs_players DataFrame
premier_clubs_players = pd.merge(premier_clubs_players, games_played_df, on='player_id', how='left')

# Fill NaN values with 0 (for players with no appearances)
premier_clubs_players['games_played'].fillna(0, inplace=True)

# Display the updated DataFrame
premier_clubs_players.head()
```

```python
# Convert 'date' column to datetime objects
appearances['date'] = pd.to_datetime(appearances['date'])

# Get a list of Premier League club IDs
premier_league_club_ids = premier_clubs['club_id'].tolist()

# Filter for events from 2023 onwards and for Premier League players
filtered_appearances = appearances[
    (appearances['date'].dt.year >= 2023) &
    (appearances['player_current_club_id'].isin(premier_league_club_ids))
]
```

```python
filtered_appearances.head()
```

```python
# Group by player_id and sum minutes played
total_minutes_played = filtered_appearances.groupby('player_id')['minutes_played'].sum().reset_index()

# Calculate total minutes played per 90 minutes
total_minutes_played['minutes_per_90'] = total_minutes_played['minutes_played'] / 90

# Rename the column for clarity
total_minutes_played.rename(columns={'minutes_played': 'total_minutes_played'}, inplace=True)
```

```python
premier_clubs_players = pd.merge(premier_clubs_players, total_minutes_played[['player_id', 'minutes_per_90']], on='player_id', how='left')
```

```python
total_minutes_played.head()
```

```python
# Group by player_id and sum goals
total_goals = filtered_appearances.groupby('player_id')['goals'].sum().reset_index()

# Merge with total_minutes_played to calculate goals per 90 minutes
goals_per_90 = pd.merge(total_goals, total_minutes_played[['player_id', 'minutes_per_90']], on='player_id', how='left')

# Calculate goals per 90 minutes
goals_per_90['goals_per_90'] = goals_per_90['goals'] / goals_per_90['minutes_per_90']

# Fill NaN values with 0 (for players with no goals)
goals_per_90['goals_per_90'].fillna(0, inplace=True)
```

```python
# Group by player_id and sum assists
total_assists = filtered_appearances.groupby('player_id')['assists'].sum().reset_index()

# Merge with total_minutes_played to calculate assists per 90 minutes
assists_per_90 = pd.merge(total_assists, total_minutes_played[['player_id', 'minutes_per_90']], on='player_id', how='left')

# Calculate assists per 90 minutes
assists_per_90['assists_per_90'] = assists_per_90['assists'] / assists_per_90['minutes_per_90']

# Fill NaN values with 0 (for players with no assists)
assists_per_90['assists_per_90'].fillna(0, inplace=True)
```

```python
premier_clubs_players = pd.merge(premier_clubs_players, goals_per_90[['player_id', 'goals_per_90']], on='player_id', how='left')
premier_clubs_players
```

```python
premier_clubs_players = pd.merge(premier_clubs_players, assists_per_90[['player_id', 'assists_per_90']], on='player_id', how='left')
premier_clubs_players
```
